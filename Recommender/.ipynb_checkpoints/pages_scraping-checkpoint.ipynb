```python
import pandas as pd
import numpy as np
import json
import time
import re #Regular expression

import requests as rq #------lib para requisi√ß√µes do servidor de hospedagem
import bs4 #BeautifulSoup4---lib para parsing do html
```

```python
queries = [ 'machine+learning', 'data+science' ]
url = 'https://www.youtube.com/results?search_query={}&sp=CAI%253D&page={}'
```

As queries 'data science' e 'machine learning' s√≥ possuem 30 pages cada

```python
for query in queries:
    for page in range(1,31):
        url_f = url.format(query, page)
        response = rq.get(url_f)
         #rq.get() faz a requisi√ß√£o do c√≥digo fonte
        search_page = response.text 
        #.text retorna o html como ums str
        with open("./raw_data/{}_{}.html".format(query, page), 'w', encoding="utf-8") as file:
        #criar a pastar raw_data manualmente, se a pasta nao existir acontece um erro
            file.write(search_page)
        time.sleep(1)
```

```python
for query in queries:
    for page in range(1, 31):
        with open("./raw_data/{}_{}.html".format(query, page), 'r', encoding="utf-8") as inp:
            html_page = inp.read()
            parsed = bs4.BeautifulSoup(html_page)
            #bs4.BeautifulSoup() processa o str do html para um obj bs4
            tags_a = parsed.findAll('a')
            #.findAll() retorna uma lista com todas as tags do parametro
            for a in tags_a:
                if a.has_attr('class'):
                #.has_attr() retorna um bool se existir(True) ou n√£o(False) o atributo da tag
                    if 'yt-uix-tile-link' in a['class']:
                    #s√≥ as tags dos videos tem 'yt-uix-tile-link'
                    #obj_bs4['attr'] retorna uma lista com os valores do attr
                        link = a['href']
                        title = a['title']
                        with open('parsed_videos.json', 'a+', encoding="utf-8") as parsed_videos:
                        #o parametro 'a+' adiciona novas linhas ao arquino
                            data = {'title':title, 'link':link, 'query':query, 'page':page}
                            parsed_videos.write('{}\n'.format( json.dumps(data) ))
                            #\n para separa cada itera√ß√£o do dic em uma linha do json
                            #.dumps() retorna o dict formata pro json
```

Os dados que eram utilizados nos pr√≥ximos passos est√£o no json

### Breve visualiza√ß√£o para ver se os dados est√£o ok

```python
df = pd.read_json('parsed_videos.json', lines=True).sort_values('title')
#lines=True diz que cada linha do json tem um linha de dados
df
```

```python
df_num = df.copy()
df_num['sum'] = 1
table = df_num.groupby('title').sum()
table
```

```python
repeated = table.loc[table['sum']>1].sort_values('sum',ascending=False)
pd.DataFrame(repeated['sum']) #20 linhas
#provavelmente por aparecerem nas duas queries (checar melhor depois)
#Curso de Intelig√™ncia Artificial com 19 repeti√ß√µes (propaganda talvez)
```
